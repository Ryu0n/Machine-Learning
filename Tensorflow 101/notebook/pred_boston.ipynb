```
import tensorflow as tf
import pandas as pd

# 과거의 데이터를 준비한다.
url = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
boston = pd.read_csv(url)
print(boston.columns)
boston.head()


# 모델을 학습시킨다.

# 모델을 이용한다.
```

```
# 독립변수, 종속변수 분리
dependent = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
independent = boston[['medv']]
print(dependent.shape, independent.shape)
```

```
# 모델의 구조를 만든다.
X = tf.keras.layers.Input(shape=[13])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```
# 모델을 학습시킨다.
model.fit(dependent, independent, epochs=500)
```

```
# 모델을 이용한다.
model.predict(dependent[0:5])
```

```
# 모델의 수식 확인
weights, bias = model.get_weights()

# 방정식 출력
func = 'y = '
for idx, weight in enumerate(weights):
  weight = str(weight[0])
  func += weight + 'x{0} + \n'.format(idx+1)

func += str(bias[0])
print(func)
```
